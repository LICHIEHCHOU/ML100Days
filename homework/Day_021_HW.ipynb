```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#請參閱作業範例 :
#Q1:畫出 cos 圖檔，並儲存
```

```python
x = np.arange(0,360)
y = np.cos(x * np.pi / 180.0)

plt.plot(x,y)
plt.show() 
plt.savefig("cos.png",dpi=300,format="png")
```

```python
#Q2:給定散點圖顏色
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.scatter(X, Y,s= 1,c='c')
plt.show()
```
