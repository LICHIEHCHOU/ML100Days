```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)

#print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import LabelEncoder
df['ç†±é–€ç¨‹åº¦'] = LabelEncoder().fit_transform(df['subscriptionCount'].values)
df = df.sort_values(by=['ç†±é–€ç¨‹åº¦'], ascending=False)
df
```
