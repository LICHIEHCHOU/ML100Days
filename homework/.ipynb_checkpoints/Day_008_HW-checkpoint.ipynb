```python
import numpy as np
```

```python
#題目
name_list = ['小明','小華','小菁','小美','小張','John','Mark','Tom']
sex_list = ['boy','boy','girl','girl','boy','boy','boy','boy']
weight_list = [67.5,75.3,50.1,45.5,80.8,90.4,78.4,70.7]
rank_list = [8,1,5,4,7,6,2,3]
myopia_list = [True,True,False,False,True,True,False,False]
```

```python
#Q1:將上列 list 依照['name', 'sex', 'weight', 'rank', 'myopia']順序擺入 array，
#   並且資料型態順序擺入[Unicode,Unicode,float,int,boolean]
```

```python
dt = np.dtype({'names':('name', 'sex', 'weight', 'rank', 'myopia'), 'formats':((np.str_, 5),(np.str_, 5), float, int, bool)})
```

```python
array = np.zeros(8, dtype=dt)
```

```python
array['name'] = name_list
array['sex'] = sex_list
array['weight'] = weight_list
array['rank'] = rank_list
array['myopia'] = myopia_list
```

```python
print(array)
```

```python
#Q2:呈上題，將 array 中體重(weight)數據集取出算出全部平均體重
```

```python
array['weight']
```

```python
np.nanmean(array['weight'])
```

```python
#Q3:呈上題，進一步算出男生(sex 欄位是 boy)平均體重、女生(sex 欄位是 girl)平均體重
```

```python
boy_average_weight = array[array['sex'] == 'boy']['weight']
boy_average_weight
```

```python
np.nanmean(boy_average_weight)
```

```python
girl_average_weight = array[array['sex'] == 'girl']['weight']
girl_average_weight
```

```python
np.nanmean(girl_average_weight)
```
