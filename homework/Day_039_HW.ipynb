###  Q1: 透過數值法計算  Age 和  Survived  是否有相關性?


```python
#將有遺失值的數值刪除
df_dropna=df_train[['Age','Survived_cate']].dropna()
```

```python
#離散 vs 連續 Eta Squared(η2)
aov = pg.anova(dv='Age', between='Survived_cate', data=df_dropna, detailed=True)
aov
```

```python
etaSq = aov.SS[0] / (aov.SS[0] + aov.SS[1])
etaSq
```

```python
def judgment_etaSq(etaSq):
    if etaSq < .01:
        qual = 'Negligible'
    elif etaSq < .06:
        qual = 'Small'
    elif etaSq < .14:
        qual = 'Medium'
    else:
        qual = 'Large'
    return(qual)
judgment_etaSq(etaSq)
```

```python
#小提琴圖示法
g = sns.catplot(x="Survived_cate", y="Age", hue="Survived_cate",
               data=df_dropna, kind="violin")
```

### Q2:透過數值法計算  Sex 和 Survived  是否有相關性?


```python
#離散 vs 離散 Cramér's V 
contTable = pd.crosstab(df_train['Sex'], df_train['Survived_cate'])
contTable
```

```python
df = min(contTable.shape[0], contTable.shape[1]) - 1
df
```

```python
crosstab, res = researchpy.crosstab(df_train['Sex'], df_train['Survived_cate'], test='chi-square')
#print(res)
print("Cramer's value is",res.loc[2,'results'])

#這邊用卡方檢定獨立性，所以採用的 test 參數為卡方 "test =" argument.
# 採用的變數在這個模組中，會自己根據資料集來判斷，Cramer's Phi if it a 2x2 table, or Cramer's V is larger than 2x2.
```

```python
#寫一個副程式判斷相關性的強度
def judgment_CramerV(df,V):
    if df == 1:
        if V < 0.10:
            qual = 'negligible'
        elif V < 0.30:
            qual = 'small'
        elif V < 0.50:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 2:
        if V < 0.07:
            qual = 'negligible'
        elif V < 0.21:
            qual = 'small'
        elif V < 0.35:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 3:
        if V < 0.06:
            qual = 'negligible'
        elif V < 0.17:
            qual = 'small'
        elif V < 0.29:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 4:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.15:
            qual = 'small'
        elif V < 0.25:
            qual = 'medium'
        else:
            qual = 'large'
    else:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.13:
            qual = 'small'
        elif V < 0.22:
            qual = 'medium'
        else:
            qual = 'large'
    return(qual)

judgment_CramerV(df,res.loc[2,'results'])
```

### Q3: 透過數值法計算  Age 和  Fare 是否有相關性?  

```python
#將有遺失值的數值刪除
df_dropna=df_train[['Age','Fare']].dropna()
```

```python
#連續 vs 連續 Pearson相關係數
# 由於 pearsonr 有兩個回傳結果，我們只需取第一個回傳值為相關係數
corr, _=stats.pearsonr(df_dropna['Age'], df_dropna['Fare'])
print(corr)
```

```python
g = sns.regplot(x="Age", y="Fare", color="g",data=df_dropna)
```
