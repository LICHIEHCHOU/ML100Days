```python
import numpy as np
```

```python
#Q1:將下兩列 array 存成 npz 檔 array1 = np.array(range(30))array2 = np.array([2,3,5])
```

```python
array1 = np.array(range(30))
array2 = np.array([2,3,5])
with open('Day_006_array.npz', 'wb') as f:
    np.savez(f, array1=array1, array2=array2)
```

```python
#Q2:讀取剛剛的 npz 檔，加入下列 array 一起存成新的 npz 檔
```

```python
array = np.array([[4,5,6],[1,2,3]])
with open('Day_006_array_new.npz', 'wb') as f:
    np.savez(f, array=array)
```
